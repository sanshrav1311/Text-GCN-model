```python
from nltk.corpus import stopwords

nltk.download("stopwords")
stop_words = stopwords.words("english")
stop_words = set(stop_words)
```

```python
print("Stop_words:", stop_words)
```

```python
def filter_tokens(tokens, stopwords):
    tokens1 = []
    for token in tokens:
        if (token not in stopwords) and (token not in [".",",",";","&","'s", ":", "?", "!","(",")",\
            "'","'m","'no","***","--","...","[","]"]):
            tokens1.append(token)
    return tokens1
```

```python
tmp_word_freq = {}
```

```python
def remove_numbers(lst):
     return [s for s in lst if any(c.isalpha() for c in s)]
```

```python
def removeStopWords(text):
    new_text = str(text).lower().translate(str.maketrans('', '', string.punctuation))
    words=nltk.word_tokenize(new_text)
    filtered_words=filter_tokens(words,stop_words)
    filtered_words=remove_numbers(filtered_words)
    doc_str = " ".join(filtered_words).strip()
    return doc_str
```

```python
def cleanKeyWords(text):
    new_text = str(text).lower().replace(";"," ").translate(str.maketrans('', '', string.punctuation))
    words=nltk.word_tokenize(new_text)
    filtered_words=filter_tokens(words,stop_words)
    filtered_words=remove_numbers(filtered_words)
    doc_str = " ".join(filtered_words).strip()
    return doc_str
```

```python
df["TITLE_CLEAN"] = df["Article title"].apply(removeStopWords)
```

```python
df["KEYWORDS_CLEAN"] = df["Article keywords"].apply(cleanKeyWords)
```

```python
df["ABSTRACT_CLEAN"] = df["Article abstract"].apply(removeStopWords)
```
